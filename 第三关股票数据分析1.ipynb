```python
import pandas as pd #导入数据分析包
from pandas_datareader import data #pandas_datareader包中可以查询雅虎的股票数据接口。
jfDict={'百度':'BIDU','IBM':'IBM','Facebook':'FB',
            '苹果':'AAPL','阿里巴巴':'BABA','中科曙光':'603019.ss'}
```

## 百度

```python
start_date = '2017-01-01'
end_date = '2018-01-01'#获取数据的时间范围
baiduDf=data.get_data_yahoo('BIDU',start_date, end_date) #get_data_yahoo表示从雅虎数据源获取股票数据
#雅虎股票数据源文档：http://pandas-datareader.readthedocs.io/en/latest/remote_data.html#yahoo-finance
```

```python
baiduDf.head()#查看前5行数据
```

```python
baiduDf.describe()#查看数据集描述统计信息
```

```python
def chan(column):
    buyPrice=column[0]
    curPrice=column[250]
    chanPrice=(curPrice-buyPrice)/buyPrice#自定义函数
    if(chanPrice>0):
        print('股票上涨： ',chanPrice)
    elif(chanPrice<0):
        print('股票下跌： ',chanPrice)
    else:#不用跟条件
        print('股票无变化')#if:elif:else:
    return chanPrice#自定义函数最后返回数据
```

```python
closeCol=baiduDf['Close']#获取百度close列数据
baiduchange=chan(closeCol)#获取Close这一列的数据,运用自定义函数
```

```python
import matplotlib.pyplot as plt#导入绘图包
plt.plot(baiduDf['Close'],color='green')#使用Close这一列数据绘制,线的颜色绿色
plt.title('BIDU')#标题
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.grid(True)#显示网格
plt.show()#显示图片    生成图片的日期格式
```
